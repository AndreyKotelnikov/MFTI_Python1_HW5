3. Следующим этапом преобразуем признак **"Опыт работы"**. Его текущий формат - это: **<Опыт работы: n лет m месяцев, периоды работы в различных компаниях…>**. 

Из столбца нам необходимо выделить общий опыт работы соискателя в месяцах, новый признак назовем "Опыт работы (месяц)"

Для начала обсудим условия решения задачи:
* Во-первых, в данном признаке есть пропуски. Условимся, что если мы встречаем пропуск, оставляем его как есть (функция-преобразование возвращает NaN)
* Во-вторых, в данном признаке есть скрытые пропуски. Для некоторых соискателей в столбце стоит значения "Не указано". Их тоже обозначим как NaN (функция-преобразование возвращает NaN)
* В-третьих, нас не интересует информация, которая описывается после указания опыта работы (периоды работы в различных компаниях)
* В-четвертых, у нас есть проблема: опыт работы может быть представлен только в годах или только месяцах. Например, можно встретить следующие варианты:
    * Опыт работы 3 года 2 месяца…
    * Опыт работы 4 года…
    * Опыт работы 11 месяцев…
    * Учитывайте эту особенность в вашем коде

Учитывайте эту особенность в вашем коде

В результате преобразования у вас должен получиться столбец, содержащий информацию о том, сколько месяцев проработал соискатель.
Выполните преобразование, ответьте на контрольные вопросы и удалите столбец **"Опыт работы"** из таблицы.


```python
# Выделим опыт работы в месяцах
# Запятой после месяца нет, поэтому будем обрезать строку по вхождению чисел и 'месяц'
def experience(x):
    if pd.isna(x) or x in ['Не указано']:
        return np.nan
    
    # Удалим информацию о периодах работы в различных компаниях
    words = x.split()
    if str.isnumeric(words[2]) and str.isnumeric(words[4]) and 'месяц' in words[5]:
        x = ' '.join(words[:6])
    elif str.isnumeric(words[2]):
        x = ' '.join(words[:4])
    else:
        raise ValueError('Unknown format')
    
    if ('год' in x or 'лет' in x) and 'месяц' in x:
        years, months = x.split()[2], x.split()[4]
        return int(years) * 12 + int(months)
    if 'год' in x or 'лет' in x:
        return int(x.split()[2]) * 12
    if 'месяц' in x:
        return int(x.split()[2])
    
    raise ValueError('Unknown format')
    
df['Опыт работы (месяц)'] = df['Опыт работы'].apply(experience)
df.drop('Опыт работы', axis=1, inplace=True)

# Чему равен медианный опыт работы (в месяцах) в нашей таблице? Ответ округлите до целых.
df['Опыт работы (месяц)'].median()
```

    Еще один важный факт: при выгрузки данных у некоторых соискателей "потерялась" информация о готовности к командировкам. Давайте по умолчанию будем считать, что такие соискатели не готовы к командировкам.
    
Выполните преобразования и удалите столбец **"Город, переезд, командировки"** из таблицы.

*Совет: обратите внимание на то, что структура текста может меняться в зависимости от указания ближайшего метро. Учите это, если будете использовать порядок слов в своей программе.*


```python
# Список городов-миллионников
million_cities = [
    'Новосибирск', 'Екатеринбург', 'Нижний Новгород', 'Казань', 'Челябинск', 
    'Омск', 'Самара', 'Ростов-на-Дону', 'Уфа', 'Красноярск', 'Пермь', 'Воронеж', 'Волгоград'
]

# Функция для извлечения города
def extract_city(text):
    text = text.split()[0]
    if 'Москва' in text:
        return 'Москва'
    elif 'Санкт-Петербург' in text:
        return 'Санкт-Петербург'
    else:
        for city in million_cities:
            if city in text:
                return city
    return 'другие'

# Функция для извлечения готовности к переезду
def extract_relocation(text: str) -> bool:
    parts = text.split(',')
    if 'готов к' in parts[1] or 'готова к' in parts[1] or 'хочу переехать' in parts[1]:
        text = parts[1].strip()
    else:
        text = parts[2].strip()
    if text.startswith('готов') or text.startswith('хочу переехать'):
        return True
    return False

# Функция для извлечения готовности к командировкам
def extract_travel(text):
    parts = text.split(',')
    last_index = len(parts) - 1
    if 'командировк' in parts[last_index]:
        text = parts[last_index].strip()
    else:
        return False
    
    if text.startswith('готов'):
        return True
    return False

# Применяем функции для извлечения данных
df['Город'] = df['Город, переезд, командировки'].apply(extract_city)
df['Готовность к переезду'] = df['Город, переезд, командировки'].apply(extract_relocation)
df['Готовность к командировкам'] = df['Город, переезд, командировки'].apply(extract_travel)
df.head(30)

# Сколько процентов соискателей живут в Санкт-Петербурге? Ответ округлите до целого.
round(df['Город'].value_counts(normalize=True)['Санкт-Петербург'] * 100)

# Сколько процентов соискателей готовы одновременно и к переездам, и к командировкам? Ответ округлите до целого.
round(df[(df['Готовность к переезду'] == True) & (df['Готовность к командировкам'] == True)].shape[0] / df.shape[0] * 100)
df.drop('Город, переезд, командировки', axis=1, inplace=True)
```

5. Рассмотрим поближе признаки **"Занятость"** и **"График"**. Сейчас признаки представляют собой набор категорий желаемой занятости (полная занятость, частичная занятость, проектная работа, волонтерство, стажировка) и желаемого графика работы (полный день, сменный график, гибкий график, удаленная работа, вахтовый метод).
На сайте hh.ru соискатель может указывать различные комбинации данных категорий, например:
* полная занятость, частичная занятость
* частичная занятость, проектная работа, волонтерство
* полный день, удаленная работа
* вахтовый метод, гибкий график, удаленная работа, полная занятость

Такой вариант признаков имеет множество различных комбинаций, а значит множество уникальных значений, что мешает анализу. Нужно это исправить!

Давайте создадим признаки-мигалки для каждой категории: если категория присутствует в списке желаемых соискателем, то в столбце на месте строки рассматриваемого соискателя ставится True, иначе - False.

Такой метод преобразования категориальных признаков называется One Hot Encoding и его схема представлена на рисунке ниже:
<img src=https://raw.githubusercontent.com/AndreyRysistov/DatasetsForPandas/main/ohe.jpg>
Выполните данное преобразование для признаков "Занятость" и "График", ответьте на контрольные вопросы, после чего удалите их из таблицы

```python
# Функция для преобразования категорий в бинарные признаки
def explode_and_get_dummies(df, column, sep):
    # Преобразуем в строки и разделяем категории
    df_exploded = df[column].astype(str).str.split(sep).explode()
    
    # Создаём дамми-переменные для взорванной серии
    dummies_exploded = pd.get_dummies(df_exploded)
    
    # Аггрегируем результаты обратно по индексу исходного DataFrame
    df_result = dummies_exploded.groupby(level=0).max()
    
    return df_result

# Преобразуем столбец "Занятость" в набор индикаторных признаков
df_busyness = explode_and_get_dummies(df, column='Занятость', sep=', ')

# Преобразуем столбец "График" в набор индикаторных признаков
df_schedule = explode_and_get_dummies(df, column='График', sep=', ')

# Объединяем новые столбцы с исходным DataFrame
df = pd.concat([df, df_busyness, df_schedule], axis=1)

df[(df['проектная работа'] == True) & (df['волонтерство'] == True)].shape[0]

df[(df['вахтовый метод'] == True) & (df['гибкий график'] == True)].shape[0]

df.drop(['Занятость', 'График'], axis=1, inplace=True)
```

4. Умножить сумму желаемой заработной платы на присоединенный курс валюты (close) и разделить на пропорцию (обратите внимание на пропуски после объединения в этих столбцах), результат занести в новый столбец "ЗП (руб)".


```python
from datetime import datetime

# Преобразует столбцы с датами в формат datetime
def custom_date_parser(x):
    return datetime.strptime(x, 'dd/MM/YY')

# Загрузим данные о курсах валют
df_currency = pd.read_csv('ExchangeRates.csv')
df_currency['date'] = pd.to_datetime(df_currency['date'], format='%d/%m/%y')

# Оставим только столбцы с курсом закрытия и валютой
df_currency = df_currency[['date', 'currency', 'close', 'proportion']]

# Преобразуем столбец "Обновление резюме" в формат datetime
df['update_date'] = pd.to_datetime(df['Обновление резюме'], format='%d.%m.%Y %H:%M')
# Оставим только дату
df['update_date'] = pd.to_datetime(df['update_date'].dt.date)

# Выделим из столбца "ЗП" сумму и валюту в отдельные столбцы
df['salary'] = df['ЗП'].apply(lambda x: int(x.split()[0]))
df['currency_main'] = df['ЗП'].apply(lambda x: x.split()[1])
df_currency
```

```python
df_currency['currency'].unique()

# Приведем названия валют к стандарту ISO
currency_dict = {
    'руб.': 'RUB',
    'бел.руб.': 'BYN',
    'EUR': 'EUR',
    'USD': 'USD',
    'KZT': 'KZT',
    'грн.': 'UAH',
    'сум': 'UZS',
    'KGS': 'KGS',
    'AZN': 'AZN',
}

df['currency_main'] = df['currency_main'].map(currency_dict)

# Присоединим к таблице с резюме таблицу с курсами валют
df = df.merge(df_currency, how='left', left_on=['update_date', 'currency_main'], right_on=['date', 'currency'])

# Заполним пропуски в столбцах close и proportion единицами
df[['close', 'proportion']] = df[['close', 'proportion']].fillna(1)

# Посчитаем ЗП в рублях
df['ЗП (руб)'] = df['salary'] * df['close'] / df['proportion']

df['ЗП (руб)'].median()

# Удалим ненужные столбцы
df.drop(['ЗП', 'salary', 'update_date', 'currency_main', 'currency', 'date', 'close', 'proportion'], axis=1, inplace=True)
df.info()
```

3. **Отличия между группами образования**: Даже при одинаковом возрасте медианные зарплаты у соискателей с высшим образованием чаще всего выше, чем у соискателей со средним или средним специальным образованием.

6. В результате анализа мы обнаружили потенциальные выбросы в признаке **"Возраст"**. Это оказались резюме людей чересчур преклонного возраста для поиска работы. Попробуйте построить распределение признака в **логарифмическом масштабе**. Добавьте к графику линии, отображающие **среднее и границы интервала метода трех сигм**. Напомним, сделать это можно с помощью метода axvline. Например, для построение линии среднего будет иметь вид:

`histplot.axvline(log_age.mean(), color='k', lw=2)`

В какую сторону асимметрично логарифмическое распределение? Напишите об этом в комментарии к графику.
Найдите выбросы с помощью метода z-отклонения и удалите их из данных, используйте логарифмический масштаб. Давайте сделаем послабление на **1 сигму** (возьмите 4 сигмы) в **правую сторону**.

Выведите таблицу с полученными выбросами и оцените, с каким возрастом соискатели попадают под категорию выбросов?

```python
# Создаем логарифмический масштаб возраста (добавляем 1, чтобы избежать логарифма от 0)
df['log_Возраст'] = np.log1p(df['Возраст'])

# Вычисляем среднее и стандартное отклонение логарифмического масштаба
log_mean = df['log_Возраст'].mean()
log_std = df['log_Возраст'].std()

# Границы метода трех сигм с послаблением на одну сигму вправо
lower_bound = log_mean - 3 * log_std
upper_bound = log_mean + 4 * log_std

# Построим гистограмму с линиями среднего и границ
plt.figure(figsize=(12, 6))
sns.histplot(df['log_Возраст'], bins=30, kde=False, edgecolor='black', alpha=0.7, label="Логарифм возраста")

# Линии для среднего и границ интервала
plt.axvline(log_mean, color='k', lw=2, label="Среднее")
plt.axvline(lower_bound, color='r', lw=2, linestyle='--', label="Нижняя граница (3 сигмы)")
plt.axvline(upper_bound, color='b', lw=2, linestyle='--', label="Верхняя граница (4 сигмы)")

plt.title('Распределение возраста в логарифмическом масштабе\nАсимметрия в правую сторону')
plt.xlabel('Логарифм возраста')
plt.ylabel('Частота')
plt.legend()
plt.show()

# Найдем выбросы с использованием z-отклонений
z_scores = (df['log_Возраст'] - log_mean) / log_std
outliers = df[(z_scores > (upper_bound - log_mean) / log_std) | (z_scores < (lower_bound - log_mean) / log_std)]
outliers
```

Выбросы включают соискателей с возрастами 15 лет (слишком молоды для работы) и 100 лет (крайне редкий возраст, вероятно ошибка ввода). \
Эти значения лучше исключить для корректного анализа.
